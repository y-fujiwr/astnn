```python
import pandas as pd
import random

df = pd.read_pickle("data/java/bcb_pair_ids_add_negative.pkl")
df2 = pd.read_csv("data/java/bcb_funcs_all.tsv", sep='\t', header=None, encoding='utf-8')
for i in range(60000):
    a = random.randint(0,59617)
    b = random.randint(0,59617)
    if a==b:
        print("lucky!")
        continue
    elif a > b:
        temp = a
        a = b
        b = temp
    df = df.append(pd.Series([df2[0][a],df2[0][b],0],index=["id1","id2","label"]),ignore_index=True)
df = df.drop_duplicates(subset=["id1","id2"])
df.to_pickle("data/java/bcb_pair_ids_balanced.pkl")
df
```
