```python
import pandas as pd
from pathlib import Path
from tqdm import tqdm,trange
import os
i=0
fileGenerator = Path("data/googlejam4_src").glob("10/*.method")
label=[]
gcj_funcs = pd.DataFrame(columns=['id','code'])
for f in tqdm(fileGenerator):
    with open(f) as method:
        gcj_funcs = gcj_funcs.append(pd.Series([i,method.read()], index=gcj_funcs.columns), ignore_index=True)
    label.append(1)
    i+=1
fileGenerator = Path("data/googlejam4_src").glob("12/*.method")
for f in tqdm(fileGenerator):
    with open(f) as method:
        gcj_funcs = gcj_funcs.append(pd.Series([i,method.read()], index=gcj_funcs.columns), ignore_index=True)
    label.append(2)
    i+=1
gcj_funcs.to_csv("data/check/check_funcs_all.csv", index=False)
gcj_funcs
```

```python
gcj_funcs.to_csv("data/check/check_funcs_all.csv")
pairs = pd.DataFrame(columns=['id1','id2','label'])
for i in trange(len(label)):
    for j in range(i+1,len(label)):
        if label[i] == label[j]:
            pairs = pairs.append(pd.Series([i,j,label[i]], index=pairs.columns),ignore_index=True)
        else:
            pairs = pairs.append(pd.Series([i,j,0], index=pairs.columns),ignore_index=True)
pairs
```
